```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Low regularization
clf = Ridge(alpha=.01)

# Load the DataFrame
df = pd.read_csv('hpi_master.csv')

df = df.dropna(subset=["index_sa"])
df = df[df["index_sa"]<450]

# One-hot encode 'place_name'
place_name_encoded = pd.get_dummies(df["place_name"], drop_first=True)


# Concatenate 'yr' and the encoded 'place_name'
X = pd.concat([df[["yr"]], place_name_encoded], axis=1)

# Define the target variable y
y = df["index_sa"]

# Train 90%, Test 10%
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.1)

# Fit the model
clf.fit(Xtrain, Ytrain)
```

```python
# Test how accurate the regression was
from sklearn.metrics import mean_squared_error, r2_score

#predicting using Xtest
prediction =  clf.predict(Xtest)

mse = mean_squared_error(prediction, Ytest)
print(mse)
```

```python
# Matplotlib visualizer
plt.scatter(Ytest, prediction)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.plot([Ytest.min(), Ytest.max()], [Ytest.min(), Ytest.max()], 'k--')  # Diagonal line
plt.show()

# Calculate R^2 value
r_squared = r2_score(Ytest, prediction)
print("R² Score:", r_squared)
```

```python

# Prepare future years DataFrame
future_years = pd.DataFrame({'yr': np.arange(2025, 2031)})

# Get the list of columns used for training the model
train_columns = clf.feature_names_in_

# Create a DataFrame with all necessary columns, filled with zeros
future_years_encoded = pd.DataFrame(0, index=np.arange(len(future_years)), columns=train_columns)

# Fill in the 'yr' column
future_years_encoded['yr'] = future_years['yr']

# Handle 'place_name' if it exists in train_columns
if any(col.startswith('place_name_') for col in train_columns):
    place_name_columns = [col for col in train_columns if col.startswith('place_name_')]
    # Assume the last place_name in your original data, or choose an appropriate one
    last_place_name = df['place_name'].iloc[-1]
    place_name_col = f'place_name_{last_place_name}'
    if place_name_col in place_name_columns:
        future_years_encoded[place_name_col] = 1

# Ensure the order of columns matches the training data
future_years_encoded = future_years_encoded[train_columns]

# Make predictions
predictions = clf.predict(future_years_encoded)

# Prepare years for plotting
years = np.concatenate([df['yr'].values, future_years['yr'].values])
index_values = np.concatenate([df['index_sa'].values, predictions])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, index_values, marker='o', label='Predicted Index Values')
plt.xlabel('Year')
plt.ylabel('Index Value')
plt.title('House Price Index Predictions')
plt.axvline(x=2024, color='red', linestyle='--', label='Prediction Start')
plt.legend()
plt.show()
```
